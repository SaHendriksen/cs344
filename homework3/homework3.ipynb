It isn't possible to create the XOR with a linear neural network. 
Instead you need to use a neural network that has two layers. The first layer contains two neurons that have the weights and thresholds of AND (threshold: >= 2) and OR (threshold: >= 1). The second layer will have the XOR which will take the inputs from the AND and OR functions and apply weights of -1 and 1 respectively to the inputs. The XOR will have a threshold of >= 1. Both biases for the different layers will have weights of 0. Using a -1 weight on the AND will account for the situation where both of the initial inputs are 1 and AND gives a result of 1.
